```python
w = 100
b = 100
print(f'w = {w}')
print(f'b = {b}')
```

```python
def compute_model_output(x, w, b):

    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)
plt.plot(x_train, tmp_f_wb, c='b', label='Our Pridiction')
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Value' )
plt.title('Houseing Price')
plt.ylabel('prince (in 1000s of doller)')
plt.xlabel('size (1000sqft)')
plt.legend()
plt.show()
```
