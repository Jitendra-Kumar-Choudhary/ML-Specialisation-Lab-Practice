```python
def zscore_normalizatin_feature(X):
    mu = np.mean(X, axis = 0)
    sigma = np.std(X, axis = 0)
    
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

```python
mu = np.mean(X_train, axis = 0)
sigma = np.std(X_train, axis = 0)
X_mean = (X_train - mu)
X_norm = (X_train - mu) / sigma 


fig, ax = plt.subplots(1, 3, figsize = (12, 3))
ax[0].scatter(X_train[:, 0], X_train[:, 3])
ax[0].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[0].set_title("unnormalized")
ax[0].axis('equal')

ax[1].scatter(X_mean[:, 0], X_mean[:, 3])
ax[1].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[1].set_title(r"X - $\mu$")
ax[1].axis('equal')

ax[2].scatter(X_norm[:, 0], X_norm[:, 3])
ax[2].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[2].set_title("Z - score normalization")
ax[2].axis('equal')

plt.tight_layout(rect = [0, 0.03, 1, 0.95] )
fig.suptitle("distribution of sub features before during after normalization")
plt.show()
```

```python
X_norm, X_mu, X_sigma = zscore_normalizatin_feature(X_train)
print(f"X_mu = {X_mu}, \nX_sigma = {X_sigma}")
print(f"peak to peak range by colomn in row X = {np.ptp(X_train, axis = 0)}")
print(f"peak by peak range by colomn in normalization X = {np.ptp(X_norm, axis = 0)}")
```
