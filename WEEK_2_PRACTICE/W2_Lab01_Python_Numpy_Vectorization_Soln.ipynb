```python
#vector indexing operations on 1-D vectors
a = np.arange(10)
print(a)

print(f"a[2].shape = {a[2].shape}, a[2] = {a[2]}, acssesing element return as scaler ")

print(f"a[-1].shape = {a[-1].shape}, a[-1] = {a[-1]}")

try:
    c = a[10]
except Exception as e:
    print("the error massage you l'll see")
    print(e)
```

```python
#vector slicing operations
a = np.arange(10)
print(f"a = {a}")

c = a[2:7:1];  print("a[2:7:1] =", c)

c = a[2:7:2];  print("a[2:7:2] =", c)

c = a[:3];  print("a[:3] =", c)

c = a[3:];  print("a[3:]=", c)

c = a[:];  print("a[:]=", c)
```

```python
#Single vector operations
a = np.array([1,2,3,4])
print(f"a              : {a}")

b = -a ;   print(f"b = -a         : {b}")

b = np.sum(a);  print(f"b = np.sum(a)  : {b}")

b = np.mean(a);  print(f"b = np.mean(a) : {b}")

b = a ** 2 ;  print(f"b = a ** 2     : {b}")
```

```python
#Vector Vector dot product
def my_dot (a,b):
       x = 0
       for i in range (x):
            x = x + a[i] * b[i]
       return x
```

```python
def my_dot(a,b):
        x=0
        for i in range(a.shape[x]):
              x = x + a[i] * b[i]
        return x
```

```python
np.random.seed(1)
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a, b)
toc = time.time()

print(f"np.dot a,b = {c :.4f}")
print(f"victorize version duration = {1000 * (toc - tic) : .4f} ms")

tic = time.time()
c = my_dot(a, b)
toc = time.time()

print(f"my_dot a,b = {c:.4f}")
print(f"loop verion duration = {1000 * (toc - tic)} ms")

del(a); del(b)
```
